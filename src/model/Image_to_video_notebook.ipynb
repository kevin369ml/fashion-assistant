```python
### Start the ComfyUI web UI, get more details regarding how to use ComfyUI at https://github.com/comfyanonymous/ComfyUI
import os, sys, time, re, subprocess
from IPython.display import clear_output

# ===== CONFIG =====
BASE = "/content"   # change if your ComfyUI path differs
PORT = 8188
os.chdir(f"{BASE}/ComfyUI")

# ===== CLEAN LOG FILE =====
open('pinggy_log.txt', 'w').close()
open('ui_log.txt', 'w').close()

clear_output(wait=True)
print("🚀 Launching ComfyUI + Pinggy Tunnel...\n")

# ===== START COMFYUI SERVER =====
ui_cmd = f"nohup python {BASE}/ComfyUI/main.py --listen 0.0.0.0 --port {PORT} > ui_log.txt 2>&1 &"
os.system(ui_cmd)
time.sleep(5)

# ===== START PINGGY TUNNEL =====
# -T disables pseudo-terminal allocation (fixes the warning)
# nohup keeps it alive even if notebook cell stops
tunnel_cmd = f"nohup ssh -T -o StrictHostKeyChecking=no -p 80 -R0:localhost:{PORT} a.pinggy.io > pinggy_log.txt 2>&1 &"
os.system(tunnel_cmd)

# ===== WAIT FOR PINGGY URL =====
print("⏳ Waiting for Pinggy public URL (up to 60s)...")

url = None
for i in range(60):  # wait up to 60 seconds
    time.sleep(2)
    if os.path.exists("pinggy_log.txt"):
        with open("pinggy_log.txt") as f:
            text = f.read()
            match = re.search(r"https?://[a-z0-9\-]+\.pinggy\.link", text)
            if match:
                url = match.group(0)
                break
    print(".", end="", flush=True)

clear_output(wait=True)
if url:
    print("✅ ComfyUI is LIVE!\n")
    print(f"🌐 Public URL: {url}")
    print("\n💡 Open this link in your browser to access the UI.")
    print("⚠️ Note: Free Pinggy tunnels expire in ~60 minutes.")
else:
    print("❌ Could not find Pinggy URL after 60 seconds.")
    print("🔍 Check the last lines of pinggy_log.txt below:")
    !tail -n 40 pinggy_log.txt
```

```python
### Could also start the web ui with Cloudflare
!wget -P ~ https://github.com/cloudflare/cloudflared/releases/latest/download/cloudflared-linux-amd64.deb
!dpkg -i ~/cloudflared-linux-amd64.deb

import subprocess
import threading
import time
import socket
import urllib.request

def iframe_thread(port):
  while True:
      time.sleep(0.5)
      sock = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
      result = sock.connect_ex(('127.0.0.1', port))
      if result == 0:
        break
      sock.close()
  print("\nComfyUI finished loading, trying to launch cloudflared (if it gets stuck here cloudflared is having issues)\n")

  p = subprocess.Popen(["cloudflared", "tunnel", "--url", "http://127.0.0.1:{}".format(port)], stdout=subprocess.PIPE, stderr=subprocess.PIPE)
  for line in p.stderr:
    l = line.decode()
    if "trycloudflare.com " in l:
      print("This is the URL to access ComfyUI:", l[l.find("http"):], end='')
    #print(l, end='')


threading.Thread(target=iframe_thread, daemon=True, args=(8188,)).start()

!python main.py --dont-print-server
```
